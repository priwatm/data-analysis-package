```python
Time = time.time()
pH.Prepare_Prices(Pairs   = Pairs,
                  Periods = Periods,
                  Nums    = Nums,
                  PrTime  = 'close',
                  PrType  = 'Ask'  )
priceHistory.Save(pH,TMPfold+'Prices_All.p')
print '----------------------'
print 'Total time %6.3f sec. '%(time.time() - Time)
```

```python
Time = time.time()
pH = priceHistory.Load(TMPfold+'Prices_All.p')
pH.Present_Prices(Period='D' ,Yield=False)
pH.Present_Prices(Period='D',Yield=True)
pH.Present_Prices(Period='M1',Yield=True)
print '----------------------'
print 'Total time %6.3f sec. '%(time.time() - Time)
```

```python
Time = time.time()
pH = priceHistory.Load(TMPfold+'Prices_All.p')
Cm = [] # Mean
Cv = [] # Var
for P in Periods:
    pS = priceStat.priceStat()
    pS.Calc_Stat(pH.Get_Prices(Pair=None,Num=None,Period=P,Yield=True))
    Cm.append(np.mean(pS.YiCorr))
    Cv.append(np.var (pS.YiCorr))
t.plot2D([Times], [Cm],None, 
         "Mean corr. coeff.", ["Time interval in min.","Value"],[True,False], 
         GrFigSize=[6,6],FileSave='./../picts/Cm.png')
t.plot2D([Times], [Cv],None, 
         "Var. corr. coeff.", ["Time interval in min.","Value"],[True,False],
         GrFigSize=[6,6],FileSave='./../picts/Cv.png')
pS.Calc_Stat(pH.Get_Prices(Pair=None,Num=None,Period='M15',Yield=True))
pS.Present_Res(FileSave='./../picts/Chist_M15.png')
pS.Calc_Stat(pH.Get_Prices(Pair=None,Num=None,Period='D',Yield=True))
pS.Present_Res(FileSave='./../picts/Chist_D.png')
print '----------------------'
print 'Total time %6.3f sec. '%(time.time() - Time)
```
