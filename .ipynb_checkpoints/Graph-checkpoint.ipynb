```python
import networkx as nx
import numpy    as np
%matplotlib inline
def Graph_Draw(Graph,Labels=None,
               Colors = {'node':'r' ,'edge':'b'     ,'font':'k'      },
               Size   = {'node':600 ,'edge':1.5     ,'font':10       },
               Shape  = {'node':'s' ,'edge':'solid' ,'font':'normal' }):
    '''
    solid|dashed|dotted,dashdo
    '''
    nx.draw         (G              = Graph,
                     pos            = nx.spring_layout(Graph),
                     with_labels    =(Labels!=None), 
                     labels         = Labels,
                     node_color     = Colors['node'],
                     edge_color     = Colors['edge'],
                     font_color     = Colors['font'],
                     node_size      = Size['node'],
                     width          = Size['edge'],
                     font_size      = Size['font'],
                     node_shape     = Shape['node'],
                     style          = Shape['edge'],
                     font_weight    = Shape['font'],
                     font_family    = 'sans-serif' )
def Parse_NetLogo(Fpath):
    f = open('./Out/World1.csv', 'r')
    while 1>0: # Read turtles coordinates
        s = f.readline()
        if "TURTLES" in s:
            f.readline()
            while 1>0:
                w = f.readline()
                if not '{all-turtles}' in w:
                    break
                w = w.split(',')
                x = w[3].split('"')[1]
                y = w[4].split('"')[1]
                print x,y
            break
    f.close()
    f = open('./Out/World1.csv', 'r')
    while 1>0:  # Read links
        s = f.readline()
        if "LINKS" in s:
            f.readline()
            while 1>0:
                w = f.readline()
                if not '"{turtle' in w:
                    break
                w = w.split(',')
                a = w[0].split(' ')[1].split('}')[0]
                b = w[1].split(' ')[1].split('}')[0]
                G.add_edge(a,b,weight=1.0)
                print a,b
            break
    f.close()
    
G=nx.Graph()
G.add_nodes_from(np.arange(10))
Labels = {}
for i in range(10):
    Labels[i] = i  
X,Y,L = Parse_NetLogo('./Out/World1.csv')
Graph_Draw(G,Labels)
```
